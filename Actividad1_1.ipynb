```python
#Importamos librerias requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("Detalle precios y productos fabricados 2022.xlsx")
df.info()
```

```python
df.head(3)
```

**FILTROS**

```python
#Filtro por objeto de los nombres de cliente que sean "PÚBLICO EN GENERAL"
filtro1=df[df["NOMBRE_CLIENTE"]=="PÚBLICO EN GENERAL"]
filtro1
```

```python
#Por filas
#Seleccionar ciertas filas y todas las columnas
filtro2=df.iloc[[1,10,15],:]
filtro2
```

```python
#Por columnas
#Seleccionar todas las filas y algunas columnas específicas
filtro3=df.iloc[:,[6,7,8]]
filtro3
```

```python
#Asignamos indice
df1=pd.read_excel("Detalle precios y productos fabricados 2022.xlsx",index_col=0)
df1.head(10)
```

```python
#Cruzar columnas y filas con base en el indice
filtro4=df1.loc[["F37044","F37048"],["DESCR","MARGEN_TOTAL_CALCULADO"]]
filtro4
```

```python
#filtro por cabecera
#Mostrar 10 primeros valores
filtro5=df1.head(10)
filtro5
```

```python
#Variables cuantitativas
#Cantidades mayores a 50
filtro6=df[df["CANT"]>50]
filtro6
```

```python
#Filtro basico con conectores logicos
#Cantidad mayor a 30 y que el nombre del vendedor sea "ALFREDO CANELA"
filtro7=df[(df["CANT"]<30) & (df["NOMBRE_VENDEDOR"]=="ALFREDO CANELA")]
filtro7
```

```python
#Filtro or
#NOMBRE_CLIENTE sea ELICAMEX o CANT sea mayor a 5
filtro8=df[(df["NOMBRE_CLIENTE"]=="ELICAMEX")|(df["CANT"]>5)]
filtro8
```

```python
#filtro not
#Muestra datos que no sean en NOMBRE_VENDEDOR=ALFREDO CANELA y que COSTO_TOTAL_CALCULADO no sea mayor a 1000
filtro9=df[~(df["NOMBRE_VENDEDOR"]=="ALFREDO CANELA") & ~(df["COSTO_TOTAL_CALCULADO"]>1000)]
filtro9
```

```python
#convertir a csv
filtro1.to_csv("Filtro1.csv")
filtro2.to_csv("Filtro2.csv")
filtro3.to_csv("Filtro3.csv")
filtro4.to_csv("Filtro4.csv")
filtro5.to_csv("Filtro5.csv")
filtro6.to_csv("Filtro6.csv")
filtro7.to_csv("Filtro7.csv")
filtro8.to_csv("Filtro8.csv")
filtro9.to_csv("Filtro9.csv")
```
