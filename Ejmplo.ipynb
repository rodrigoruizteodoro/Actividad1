**FILTROS**

```python
#Por objeto
filtro1=df[df["NOMBRE"]=="DONGA HWASUNG MEXICO"]
filtro1
```

```python
#Por filas
#Seleccionar ciertas filas y todas las columnas
filtro2=df.iloc[[0,24,100],:]
filtro2
```

```python
#Por columnas
#Seleccionar todas las filas y algunas columnas
filtro3=df.iloc[:,[0,3]]
filtro3
```

```python
#Asignamos indice
df1=pd.read_excel("Antigüedad de saldos 17.03.2023.xlsx",index_col=1)
df1
```

```python
#Cruzar columnas y filas con base en el indice
filtro4=df1.loc[["HELVEX","DONGA HWASUNG MEXICO"],["MONTO ADEUDADO"]]
filtro4
```

```python
#filtro por cabecera
filtro5=df.head(10)
filtro5
```

```python
#Variables cuantitativas
filtro6=df[df["MONTO ADEUDADO"]>70000]
filtro6
```

```python
#Filtro basico con conectores logicos
filtro7=df[(df["MONTO ADEUDADO"]<6000) & (df["FECHA_FACTURA"]=="2023-03-14")]
filtro7
```

```python
filtro8=df[(df["MONTO ADEUDADO"]<6000)| (df["FECHA_FACTURA"]=="2023-03-14")]
filtro8
```

```python
#filtro not
filtro9=df[~(df["MONTO ADEUDADO"]<6000) & ~(df["FECHA_FACTURA"]=="2023-03-14")]
filtro9
```

```python
#convertir a csv
filtro9.to_csv("Entregable1.csv")
```
