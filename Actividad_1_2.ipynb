```python
#Importamos librerias requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("Datos de Facturación.xlsx",index_col=4)
df.head(5)
```

```python
df.info()
```

```python
#Se convierte el dato a int
df.CVE_CLPV=df.CVE_CLPV.replace({"o":"0"},regex=False)
df.CVE_CLPV=df.CVE_CLPV.replace({"O":"0"},regex=False)
df.CVE_CLPV=df.CVE_CLPV.replace({"6781|":"6781"},regex=False)
df.CVE_CLPV=df.CVE_CLPV.replace({"3612C":"3612"},regex=False)
df.CVE_CLPV=df.CVE_CLPV.replace({"'":"0"},regex=False)
df.CVE_CLPV=df.CVE_CLPV.replace({"      ":"0"},regex=False)
df.CVE_CLPV=df.CVE_CLPV.replace({"2385A":"2385"},regex=False)
df.CVE_CLPV=df.CVE_CLPV.replace({"2510A":"2510"},regex=False)
df.CVE_CLPV=df.CVE_CLPV.replace({"8440Q":"8440"},regex=False)
df.CVE_CLPV=df.CVE_CLPV.replace({"M":"0"},regex=False)
```

```python
#Valor de 1000 a 2000
#CVE_CLPV
df["CVE_CLPV"]=df["CVE_CLPV"].astype(float)
filtro1=df[(df["CVE_CLPV"]<2000) & (df["CVE_CLPV"]>1000)]
filtro1
```

```python
#CVE_VEND
#Todas las claves excepto “5” y “4”)
filtro2=df[~(df["CVE_VEND"]==4) & ~(df["CVE_VEND"]==5)]
filtro2
```

```python
#FECHA_ENT
#Fechas de “28/02/2022"
filtro3=df[df['FECHA_ENT'] =='2022-04-13 00:00:00']
filtro3
```

```python
#CAN_TOT
#Cantidades menores a 5951.7)   o  STATUS (“E”)

filtro4=df[(df["CAN_TOT"]<5951.7)| (df["CAN_TOT"]=="E")]
filtro4
```

```python
#Solo las columnas: CVE_DOC, FECHA_ENT, FECHA_VEN y CAN_TOT
filtro5=df.loc[:,["CVE_DOC","FECHA_ENT","FECHA_VEN","CAN_TOT"]]
filtro5
```

```python
#Solo las filas de 7001-7099
filtro6=df.iloc[7001:70100,:]
filtro6
```

```python
#Index= CVE_VEND (valores: 1, 2)
#(columna: FECHAELAB)
df1=pd.read_excel("Datos de Facturación.xlsx",index_col=3)
df1.head(3)
```

```python
#Index= CVE_VEND (valores: 1, 2)  (columna: FECHAELAB)
filtro7=df1.loc[[1,1],["FECHAELAB"]]
filtro7
```

```python
#convertir a csv
filtro1.to_csv("Filtro1.2.csv")
filtro2.to_csv("Filtro2.2.csv")
filtro3.to_csv("Filtro3.2.csv")
filtro4.to_csv("Filtro4.2.csv")
filtro5.to_csv("Filtro5.2.csv")
filtro6.to_csv("Filtro6.2.csv")
filtro7.to_csv("Filtro7.2.csv")
```
