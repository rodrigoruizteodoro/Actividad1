```python
#Eliminar la fila “0” de “Unnamed” y Eliminar la columna “0” y la columna “Referencia”
df=pd.read_excel("Inmuebles.xlsx",skiprows=1)
df.head(5)
```

```python
a=df.drop(['Referencia'],axis=1)
df1=a
df1
```

```python
#b) Todos las registros de “Fecha Alta : igual a 2007-04-29 00:00:00 y a 2004-01-02 00:00:00
b=df1[(df1["Fecha_Alta"]=="2007-04-29 00:00:00") | (df1["Fecha_Alta"]=="2004-01-02 00:00:00")]
b
```

```python
#c) Todos los registros de “ Tipo: parking, local, industrial y piso”
c=df1[(df1["Tipo"]=="Parking") | (df1["Tipo"]=="Local")|(df1["Tipo"]=="Industrial")|(df1["Tipo"]=="Piso")]
c
```

```python
#d) Todos los registros con “Provincia” igual a “Girona” y con “Superficie” mayor a “100”
d=df1[(df1["Provincia"]=="Girona") & (df1["Superficie"]>100)]
d
```

```python
#e) Todos los registros con “Precio Venta: mayor a 1000,000” o “vendedor” igual a “Carmen o Joaquin”
e=df1[(df1["Precio Venta"]>1000000) | (df1["Vendedor"]=="Carmen")|(df1["Vendedor"]=="Joaquín")]
e
```

```python
#f) Todos los registros que no sean de “Provincia : Barcelona“
f=df1[~(df1["Provincia"]=="Barcelona")]
f
```

```python
#Todos los registros que no sean de “Operación: alquiler “ ni “Tipo: oficina“
g=df1[~(df1["Operación"]=="Alquiler")& ~(df1["Tipo"]=="Oficina")]
g
```
